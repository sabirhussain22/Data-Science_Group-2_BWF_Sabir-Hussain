# Task 13: Data manipulation with Pandas (indexing, selection, grouping)

```python
import pandas as pd
```

#### Load a DataFrame from a CSV file. Display the first and last five rows of the DataFrame.

```python
file_path = 'laptop_price_data.csv'
df = pd.read_csv(file_path)
df
```

```python
df.head()
```

```python
df.tail()
```

#### Set a specific column as the index of the DataFrame.

```python
df.set_index('Company', inplace=True)
```

```python
df.head()
```

```python
#### Select a specific column and display its values.
```

```python
ram_values = df['Ram']
ram_values
```

#### Select multiple columns and display the resulting DataFrame.

```python
selected_columns_df = df[['TypeName', 'Price', 'Os']]
selected_columns_df
```

#### Select a subset of rows using the .loc method.

```python
subset_loc = df.loc['Apple']
subset_loc
```

#### Select a subset of rows and columns using the .iloc method.

```python
subset_iloc = df.iloc[0:5, 0:3]
subset_iloc
```

#### Filter rows based on a condition.

```python
filtered_rows = df[df['Price'] > 11]
filtered_rows
```

#### Group the DataFrame by a specific column and calculate the mean of each group.

```python
mean_by_company = df.groupby('Company').mean(numeric_only=True)
mean_by_company
```

#### Group the DataFrame by multiple columns and calculate the sum of each group.

```python
sum_by_company_type = df.groupby(['Company', 'TypeName']).sum(numeric_only=True)
sum_by_company_type
```

#### Use the agg method to apply multiple aggregation functions to grouped data.

```python
agg_functions = df.groupby('Company').agg({'Price': ['mean', 'max'], 'Ram': ['mean', 'min']})
agg_functions
```

#### Calculate the size of each group.

```python
size_by_company = df.groupby('Company').size()
size_by_company
```

#### Select rows based on multiple conditions.

```python
multiple_conditions = df[(df['Ram'] > 8) & (df['TouchScreen'] == 1)]
multiple_conditions
```

#### Use the query method to filter rows.

```python
query_filtered = df.query('Price > 11')
query_filtered
```

#### Use isin to filter rows based on a list of values.

```python
isin_filtered = df[df['Os'].isin(['Windows', 'Mac'])]
isin_filtered
```

#### Select specific columns and rename them.

```python
renamed_columns_df = df[['Ram', 'Weight', 'Price']].rename(columns={'Ram': 'Memory', 'Weight': 'Mass', 'Price': 'Cost'})
renamed_columns_df
```
