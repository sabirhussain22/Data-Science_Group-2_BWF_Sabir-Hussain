### Subplots

Subplots allow multiple plots to be displayed in a single figure, facilitating the comparison of different data visualizations. By organizing multiple plots in a grid layout, subplots provide a comprehensive view of various aspects of the data within one cohesive figure.

- **2x2 Subplots**: A common configuration is a 2x2 grid of subplots, which can include different plot types such as line plots, scatter plots, bar plots, and histograms.
- **Customization**: Each subplot can be individually customized with titles, axis labels, legends, and different plot styles to enhance clarity and readability.
- **Applications**: Subplots are particularly useful for comparing trends, distributions, and relationships across multiple variables or datasets.


<a id=11></a>
# <span style='color:#ffffff; font-size:70%;'> Generate a figure with 2x2 subplots, each containing a different type of plot (e.g., line plot, scatter plot, bar plot, and histogram).</span>

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

age_counts = titanic_df['Age'].value_counts().sort_index()
axes[0, 0].plot(age_counts.index, age_counts.values, color='b')

axes[0, 1].scatter(titanic_df['Age'], titanic_df['Fare'], color='r')

survival_counts = titanic_df['Survived'].value_counts()
axes[1, 0].bar(survival_counts.index, survival_counts.values, color=['skyblue', 'lightgreen'])

axes[1, 1].hist(titanic_df['Age'].dropna(), bins=20, color='purple', edgecolor='black')

plt.tight_layout()
plt.show()
```

<a id=12></a>
# <span style='color:#ffffff; font-size:70%;'> Customize the subplots with titles, axis labels, and legends.</span>

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

age_counts = titanic_df['Age'].value_counts().sort_index()
axes[0, 0].plot(age_counts.index, age_counts.values, color='b')
axes[0, 0].set_title('Line Plot: Age vs Fare')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Fare')

axes[0, 1].scatter(titanic_df['Age'], titanic_df['Fare'], color='r')
axes[0, 1].set_title('Scatter Plot: Age vs Fare')
axes[0, 1].set_xlabel('Age')
axes[0, 1].set_ylabel('Fare')

survival_counts = titanic_df['Survived'].value_counts()
axes[1, 0].bar(survival_counts.index, survival_counts.values, color=['skyblue', 'lightgreen'])
axes[1, 0].set_title('Bar Plot: Survival Counts')
axes[1, 0].set_xlabel('Survived')
axes[1, 0].set_ylabel('Count')

axes[1, 1].hist(titanic_df['Age'].dropna(), bins=20, color='purple', edgecolor='black')
axes[1, 1].set_title('Histogram: Age Distribution')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Histograms

Histograms are used to visualize the distribution of a numerical dataset by dividing the data into bins and counting the number of observations in each bin. They provide a clear view of the dataâ€™s underlying distribution, including its central tendency, variability, and shape.

- **Customization**: The number of bins can be adjusted to provide a more detailed or broader view of the data distribution. Histograms can also be customized with different colors, labels, and titles.
- **Multiple Histograms**: Plotting multiple histograms on the same figure allows for the comparison of different datasets, making it easy to identify similarities and differences in their distributions.
- **Normalization**: Normalizing histograms ensures that the area under each histogram sums to one, which is useful for comparing datasets with different sample sizes.


<a id=21></a>
# <span style='color:#ffffff; font-size:70%;'>Create a histogram for a given numerical data set. Customize the number of bins.
</span>

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic_df['Fare'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Histogram: Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

<a id=22></a>
# <span style='color:#ffffff; font-size:70%;'> Plot multiple histograms on the same figure to compare different data sets.</span>

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic_df[titanic_df['Survived'] == 0]['Age'].dropna(), bins=20, alpha=0.5, label='Did Not Survive', color='red', edgecolor='black')
plt.hist(titanic_df[titanic_df['Survived'] == 1]['Age'].dropna(), bins=20, alpha=0.5, label='Survived', color='green', edgecolor='blue')
plt.title('Histograms: Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

<a id=23></a>
# <span style='color:#ffffff; font-size:70%;'> Normalize the histograms and add appropriate labels and titles.</span>

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic_df[titanic_df['Survived'] == 0]['Age'].dropna(), bins=20, alpha=0.5, density=True, label='Did Not Survive', color='red', edgecolor='black')
plt.hist(titanic_df[titanic_df['Survived'] == 1]['Age'].dropna(), bins=20, alpha=0.5, density=True, label='Survived', color='green', edgecolor='blue')
plt.title('Normalized Histograms: Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### Box Plots

Box plots, also known as box-and-whisker plots, summarize the distribution of a dataset by displaying its median, quartiles, and outliers. They provide a concise visual representation of the dataâ€™s spread and identify any potential outliers.

- **Single Box Plot**: A single box plot shows the distribution of one dataset, highlighting its central tendency and variability.
- **Multiple Box Plots**: Multiple box plots can be displayed side by side in a single figure to compare the distributions of different datasets.
- **Customization**: Box plots can be customized with titles, labels, and different colors to enhance their visual appeal and make comparisons easier.


<a id=31></a>
# <span style='color:#ffffff; font-size:70%;'> Create a box plot for a given numerical data set</span>

```python
plt.figure(figsize=(10, 6))
plt.boxplot(titanic_df['Age'].dropna())
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

<a id=32></a>
# <span style='color:#ffffff; font-size:70%;'> Generate multiple box plots in a single figure to compare different data sets.</span>

```python
plt.figure(figsize=(12, 8))
plt.boxplot([titanic_df['Age'].dropna(), titanic_df['Fare'].dropna()], labels=['Age', 'Fare'])
plt.title('Box Plots of Age and Fare')
plt.ylabel('Values')
plt.show()
```

```python
pclass_1 = titanic_df[titanic_df['Pclass'] == 1]
pclass_2 = titanic_df[titanic_df['Pclass'] == 2]
pclass_3 = titanic_df[titanic_df['Pclass'] == 3]
```

```python
plt.figure(figsize=(18, 10))

plt.subplot(2, 3, 1)
plt.boxplot(pclass_1['Age'].dropna(), patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Age of Pclass 1')
plt.ylabel('Age')

plt.subplot(2, 3, 2)
plt.boxplot(pclass_2['Age'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Age of Pclass 2')
plt.ylabel('Age')

plt.subplot(2, 3, 3)
plt.boxplot(pclass_3['Age'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightcoral'))
plt.title('Age of Pclass 3')
plt.ylabel('Age')

plt.subplot(2, 3, 4)
plt.boxplot(pclass_1['Fare'].dropna(), patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Fare of Pclass 1')
plt.ylabel('Fare')

plt.subplot(2, 3, 5)
plt.boxplot(pclass_2['Fare'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Fare of Pclass 2')
plt.ylabel('Fare')

plt.subplot(2, 3, 6)
plt.boxplot(pclass_3['Fare'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightcoral'))
plt.title('Fare of Pclass 3')
plt.ylabel('Fare')

plt.suptitle('Box Plots of Age and Fare by Pclass')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

<a id=33></a>
# <span style='color:#ffffff; font-size:70%;'> Customize the box plots with titles, labels, and change the colors.
</span>

```python
plt.figure(figsize=(12, 8))
boxprops = dict(linestyle='-', linewidth=2, color='blue')
medianprops = dict(linestyle='--', linewidth=2, color='red')
meanprops = dict(marker='o', markerfacecolor='green', markeredgecolor='black', markersize=10)

plt.boxplot([titanic_df['Age'].dropna(), titanic_df['Fare'].dropna()], 
            labels=['Age', 'Fare'], 
            boxprops=boxprops, 
            medianprops=medianprops, 
            meanprops=meanprops, 
            showmeans=True)
plt.title('Customized Box Plots of Age and Fare')
plt.ylabel('Values')
plt.show()
```
