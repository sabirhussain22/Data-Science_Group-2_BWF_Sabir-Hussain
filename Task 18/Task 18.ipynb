## Introduction to Seaborn
Seaborn is a powerful and versatile data visualization library in Python, built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn is particularly well-suited for visualizing complex datasets and comes with several built-in themes that improve the aesthetics of your plots.

### Key Features of Seaborn
- Built-in Themes: Seaborn offers several built-in themes to improve the overall look of your visualizations.
- Color Palettes: Easily apply color palettes to enhance the visual appeal of your plots.
- Statistical Plots: Create complex statistical plots with simple commands.
- Integration with Pandas: Seamlessly integrate with Pandas DataFrames to create plots directly from data frames.
- Customizable Plots: Customize every aspect of your plots to suit your needs.

### Scatter Plot
Scatter plots are a fundamental tool for visualizing the relationship between two continuous variables. They are particularly useful for identifying patterns, trends, and potential correlations within your data. In Seaborn, scatter plots can be easily created and customized, making it an excellent choice for exploratory data analysis.

#### Key Features of Scatter Plots
- Relationship Visualization: Scatter plots show how one variable is affected by another.
- Trend Identification: They help in identifying trends, clusters, and outliers.
- Customization: Seaborn allows extensive customization including color, size, and style of the points.

```python
scatter1 = sns.scatterplot(x='age', y='chol', hue='sex', data=heart_disease)
scatter1.set(title='Age vs Cholesterol by Sex', xlabel='Age', ylabel='Cholesterol')
scatter1.figure.show()
```

```python
scatter2 = sns.scatterplot(x='age', y='thalach', hue='sex', data=heart_disease)
scatter2.set(title='Age vs Maximum Heart Rate by Sex', xlabel='Age', ylabel='Maximum Heart Rate')
scatter2.figure.show()
```

```python
scatter3 = sns.scatterplot(x='Fare', y='Age', hue='Pclass', data=titanic)
scatter3.set(title='Fare vs Age by Class', xlabel='Fare', ylabel='Age')
scatter3.figure.show()
```

### Line Plot
Line plots are an essential tool for visualizing data trends over a continuous interval or time period. They are particularly useful for examining how data points change over time or in response to other continuous variables. Seaborn makes it straightforward to create and customize line plots, providing clear insights into your data.
#### Key Features of Line Plots
- Trend Visualization: Line plots help in identifying trends and patterns over a continuous range.
- Comparative Analysis: Multiple lines can be plotted to compare different groups or categories.
- Customization: Seaborn allows extensive customization of line plots, including color, style, and markers.

```python
line1 = sns.lineplot(x='age', y='chol', hue='sex', data=heart_disease)
line1.set(title='Age vs Cholesterol by Sex', xlabel='Age', ylabel='Cholesterol')
line1.figure.show()
```

```python
line2 = sns.lineplot(x='age', y='thalach', hue='sex', data=heart_disease)
line2.set(title='Age vs Maximum Heart Rate by Sex', xlabel='Age', ylabel='Maximum Heart Rate')
line2.figure.show()
```

```python
line3 = sns.lineplot(x='Age', y='Fare', hue='Pclass', data=titanic)
line3.set(title='Age vs Fare by Class', xlabel='Age', ylabel='Fare')
line3.figure.show()
```

### Bar Plot
Bar plots are a fundamental tool for visualizing the distribution and comparison of categorical data. They represent data with rectangular bars, where the length or height of each bar corresponds to the value of the variable. Seaborn simplifies the creation and customization of bar plots, making it easy to convey meaningful insights.

#### Key Features of Bar Plots
Categorical Data Representation: Bar plots are ideal for visualizing and comparing the quantities of categorical data.
Easy Comparison: They make it straightforward to compare different groups or categories.
Customization: Seaborn allows extensive customization of bar plots, including color, orientation, and aggregation functions.

```python
bar1 = sns.barplot(x='sex', y='chol', data=heart_disease)
bar1.set(title='Cholesterol by Sex', xlabel='Sex', ylabel='Cholesterol')
bar1.figure.show()
```

```python
bar2 = sns.barplot(x='cp', y='thalach', data=heart_disease)
bar2.set(title='Maximum Heart Rate by Chest Pain Type', xlabel='Chest Pain Type', ylabel='Maximum Heart Rate')
bar2.figure.show()
```

```python
bar3 = sns.barplot(x='Pclass', y='Fare', data=titanic)
bar3.set(title='Fare by Class', xlabel='Class', ylabel='Fare')
bar3.figure.show()
```

```python
bar4 = sns.barplot(x='Embarked', y='Age', data=titanic)
bar4.set(title='Age by Embarked', xlabel='Embarkation', ylabel='Age')
bar4.figure.show()
```

### count Plot
Count plots are a specific type of bar plot used to visualize the count of observations in each categorical bin. They are useful for understanding the distribution of categorical data and identifying the frequency of different categories. Seaborn provides a simple and effective way to create and customize count plots.

#### Key Features of Count Plots
- Frequency Visualization: Count plots are ideal for visualizing the frequency of different categories in a dataset.
- Easy Comparison: They make it straightforward to compare the counts of different groups or categories.
- Customization: Seaborn allows extensive customization of count plots, including color, orientation, and grouping.

```python
count1 = sns.countplot(x='sex', data=heart_disease)
count1.set(title='Count of Sex', xlabel='Sex', ylabel='Count')
count1.figure.show()
```

```python
count2 = sns.countplot(x='cp', data=heart_disease)
count2.set(title='Count of Chest Pain Type', xlabel='Chest Pain Type', ylabel='Count')
count2.figure.show()
```

```python
count3 = sns.countplot(x='Pclass', data=titanic)
count3.set(title='Count of Class', xlabel='Class', ylabel='Count')
count3.figure.show()
```

```python
count4 = sns.countplot(x='Embarked', data=titanic)
count4.set(title='Count of Embarkaed', xlabel='Embarkation', ylabel='Count')
count4.figure.show()
```

### Box Plot
Box plots (or box-and-whisker plots) are a powerful tool for summarizing the distribution of a dataset. They display the data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. Box plots are particularly useful for identifying outliers and understanding the spread and skewness of the data.

#### Key Features of Box Plots
- Distribution Visualization: Box plots provide a clear summary of the distribution of data.
- Outlier Detection: They help in identifying outliers within the dataset.
- Comparative Analysis: Box plots can compare distributions across different groups or categories.
- Customization: Seaborn allows extensive customization of box plots, including color, orientation, and grouping.

```python
box1 = sns.boxplot(x='sex', y='chol', data=heart_disease)
box1.set(title='Cholesterol by Sex', xlabel='Sex', ylabel='Cholesterol')
box1.figure.show()
```

```python
box2 = sns.boxplot(x='cp', y='trestbps', data=heart_disease)
box2.set(title='Resting Blood Pressure by Chest Pain Type', xlabel='Chest Pain Type', ylabel='Resting Blood Pressure')
box2.figure.show()
```

```python
box3 = sns.boxplot(x='Pclass', y='Age', data=titanic)
box3.set(title='Age by Class', xlabel='Class', ylabel='Age')
box3.figure.show()
```

# <b><span style='color:#ff6200'>Customization</span>

```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.scatterplot(x='age', y='chol', hue = 'sex', data=heart_disease, ax=axes[0], palette='husl')
axes[0].set_title('Age vs Cholesterol (whitegrid) Palette(husl)')
sns.scatterplot(x='Fare', y='Age', hue='Pclass', data=titanic, ax=axes[1], palette='Paired')
axes[1].set_title('Fare vs Age (whitegrid) Palette(Paired)')

plt.show()

sns.set_style('darkgrid')
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.scatterplot(x='age', y='chol', hue = 'sex', data=heart_disease, ax=axes[0], palette='viridis')
axes[0].set_title('Age vs Cholesterol (darkgrid) Palette(viridis)')
sns.scatterplot(x='Fare', y='Age', hue = 'Pclass', data=titanic, ax=axes[1], palette='Set1')
axes[1].set_title('Fare vs Age (darkgrid) Palette(Set1)')

plt.show()
```

```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.lineplot(x='age', y='chol', hue = 'sex', data=heart_disease, ax=axes[0], palette='husl')
axes[0].set_title('Age vs Cholesterol (whitegrid) Palette (husl)')
sns.lineplot(x='Age', y='Fare', hue = 'Pclass', data=titanic, ax=axes[1], palette='Paired')
axes[1].set_title('Fare vs Age (whitegrid) Palette (Paired)')

plt.show()

sns.set_style('darkgrid')
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.lineplot(x='age', y='chol', data=heart_disease, ax=axes[0], palette='viridis')
axes[0].set_title('Age vs Cholesterol (darkgrid) Palette (viridis)')
sns.lineplot(x='Age', y='Fare', hue='Pclass', data=titanic, ax=axes[1], palette='Set1')
axes[1].set_title('Fare vs Age (darkgrid) Palette (Set1)')

plt.show()
```

```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(x='cp', y='trestbps', hue= 'sex', data=heart_disease, ax=axes[0], palette='Set3')
axes[0].set_title('Resting Blood Pressure by Chest Pain Type (whitegrid) Palette (Set3)')
sns.boxplot(x='Pclass', y='Age', hue = 'Sex',data=titanic, ax=axes[1], palette='magma')
axes[1].set_title('Age by Class (whitegrid) Palette (magma)')

plt.show()

sns.set_style('darkgrid')
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(x='cp', y='trestbps', hue= 'sex', data=heart_disease, ax=axes[0], palette='viridis')
axes[0].set_title('Cholesterol by Sex (darkgrid) Palette (viridis)')
sns.boxplot(x='Pclass', y='Age', hue= 'Sex', data=titanic, ax=axes[1], palette='Set1')
axes[1].set_title('Age by Class (darkgrid) Palette (Set1)')

plt.show()
```
