## Basic Plots

Basic plots are fundamental visualizations that provide a clear and straightforward way to present data trends and patterns over time or across different categories. They are essential tools for initial data exploration and communication of key insights.

### Types of Basic Plots

1. **Line Plots**:
   - Line plots display data points connected by straight lines. They are ideal for visualizing trends over time or continuous data.

2. **Multiple Line Plots**:
   - Multiple line plots show more than one data series on the same graph, allowing for the comparison of different datasets.

### Customization Options

Basic plots offer various customization options to enhance their clarity and visual appeal:

- **Line Styles and Colors**: Line styles (solid, dashed, dotted) and colors can be customized to differentiate between multiple data series.
- **Markers**: Markers can be added to line plots to highlight individual data points. Their styles, sizes, and colors can be adjusted.
- **Grid Lines**: Grid lines can be included to enhance the precision and readability of the plots.
- **Legends**: Legends help identify different data series or categories, making the plots more informative.


<a id=11></a>
# <span style='color:#ffffff; font-size:70%;'> Create a simple line plot.</span>

```python
age_counts = titanic_df['Age'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(age_counts.index, age_counts.values)
plt.title('Number of Passengers by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

<a id=12></a>
# <span style='color:#ffffff; font-size:70%;'> Plot multiple lines on the same graph.</span>

```python
plt.figure(figsize=(10, 6))
plt.plot(titanic_df.index, titanic_df['Age'], label='Age')
plt.plot(titanic_df.index, titanic_df['Fare'], label='Fare')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Age and Fare')
plt.legend()
plt.show()
```

<a id=13></a>
# <span style='color:#ffffff; font-size:70%;'> Add markers to a line plot.</span>

```python
age_counts = titanic_df['Age'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(age_counts.index, age_counts.values, marker='o')
plt.title('Number of Passengers by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(titanic_df.index, titanic_df['Age'], label='Age', marker='o')
plt.plot(titanic_df.index, titanic_df['Fare'], label='Fare', marker='o')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Age and Fare')
plt.legend()
plt.show()
```

<a id=14></a>
# <span style='color:#ffffff; font-size:70%;'> Customize line styles and colors.</span>

```python
age_counts = titanic_df['Age'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(age_counts.index, age_counts.values, marker='o', linestyle='--', color='black')
plt.title('Number of Passengers by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(titanic_df.index, titanic_df['Age'], label='Age', linestyle='-', color='red')
plt.plot(titanic_df.index, titanic_df['Fare'], label='Fare', linestyle='-.', color='green')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Age and Fare')
plt.legend()
plt.show()
```

## Scatter Plots

Scatter plots are a powerful tool for visualizing the relationship between two continuous variables. By displaying data points on a two-dimensional plane, scatter plots help in identifying correlations, patterns, and outliers within the data.

### Features of Scatter Plots

1. **Correlation Identification**:
   - Scatter plots are useful for identifying the type and strength of the relationship between two variables (positive, negative, or no correlation).

2. **Pattern Recognition**:
   - Patterns such as clusters, trends, and deviations can be easily observed.

3. **Outlier Detection**:
   - Outliers, or data points that deviate significantly from the overall pattern, are easily spotted in scatter plots.

### Customization Options

Scatter plots offer various customization options to enhance the visual representation of data:

- **Marker Styles**: The style of markers (e.g., circles, squares, triangles) can be customized to differentiate between different data series.
- **Marker Sizes**: Marker size can be adjusted to represent an additional variable, adding a third dimension to the plot.
- **Marker Colors**: Different colors can be used to distinguish between categories or groups within the data.
- **Transparency**: Marker transparency (alpha) can be adjusted to improve the visibility of overlapping points.
- **Labels and Titles**: Axis labels, titles, and data point labels can be added to provide context and clarity.
- **Grid Lines**: Grid lines can be included to enhance readability and precision in interpreting data points.



<a id=21></a>
# <span style='color:#ffffff; font-size:70%;'>Create a basic scatter plot.</span>

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic_df['Age'], titanic_df['Fare'], label='Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot: Fare vs. Age')
plt.show()
```

<a id=22></a>
# <span style='color:#ffffff; font-size:70%;'> Customize marker styles, sizes, and colors.</span>

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic_df['Age'], titanic_df['Fare'], marker='*', c='red', label='Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Customized Scatter Plot: Fare vs. Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic_df['Age'], titanic_df['Fare'], s=100, marker='^', c='black', label='Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Customized Scatter Plot: Fare vs. Age')
plt.show()
```

<a id=23></a>
# <span style='color:#ffffff; font-size:70%;'> Plot multiple scatter plots on the same graph.</span>

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic_df.index, titanic_df['Age'], marker='^', c='blue', label='Age')
plt.scatter(titanic_df.index, titanic_df['Fare'], marker='x', c='green', label='Fare')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Scatter Plots: Age and Fare')
plt.show()
```

<a id=24></a>
# <span style='color:#ffffff; font-size:70%;'> Add labels and titles to the scatter plot.</span>

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic_df['Age'], titanic_df['Fare'], marker='o', c='purple', label='Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot with Labels and Titles: Fare vs. Age')
plt.legend()
plt.grid(True)
plt.show()
```

## Bar Plots

Bar plots are an effective way to compare quantities across different categories. They use rectangular bars to represent data values, with the length or height of each bar corresponding to the value it represents. Bar plots can be oriented either vertically or horizontally, and they are particularly useful for visualizing categorical data.

### Types of Bar Plots

1. **Vertical Bar Charts**:
   - Vertical bar charts, also known as column charts, display data with vertical bars extending from the horizontal axis.
   - They are ideal for comparing values across different categories or groups.

2. **Horizontal Bar Charts**:
   - Horizontal bar charts use bars that extend from the vertical axis.
   - They are especially useful when category names are long or when you have many categories, as they make the chart easier to read.

### Customization Options

Bar plots offer a variety of customization options to enhance their clarity and visual appeal:

- **Bar Colors**: Colors can be customized to distinguish between different categories or to highlight specific bars.
- **Edge Colors**: Adding edge colors to bars can improve the visual distinction between adjacent bars.
- **Bar Width**: The width of bars can be adjusted to accommodate the number of categories and the overall size of the plot.
- **Labels**: Category labels and data value labels can be added to provide more information and context.
- **Grid Lines**: Adding grid lines can help in accurately reading the values represented by the bars.
- **Annotations**: Annotations can be used to highlight specific data points or to add additional information directly onto the plot.

<a id=31></a>
# <span style='color:#ffffff; font-size:70%;'> Create a vertical bar chart.</span>

```python
plt.figure(figsize=(10, 6))
titanic_df['Survived'].value_counts().plot(kind='bar')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Vertical Bar Chart: Survival Count')
plt.show()
```

<a id=32></a>
# <span style='color:#ffffff; font-size:70%;'> Create a horizontal bar chart.</span>

```python
plt.figure(figsize=(10, 6))
titanic_df['Survived'].value_counts().plot(kind='barh')
plt.xlabel('Count')
plt.ylabel('Survived')
plt.title('Horizontal Bar Chart: Survival Count')
plt.show()
```

<a id=33></a>
# <span style='color:#ffffff; font-size:70%;'> Customize bar colors and add edge colors.</span>

```python
plt.figure(figsize=(10, 6))
titanic_df['Survived'].value_counts().plot(kind='bar', color='green', edgecolor='red')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Customized Vertical Bar Chart: Survival Count')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
titanic_df['Pclass'].value_counts().plot(kind='barh', color='purple', edgecolor='black')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Customized Vertical Bar Chart: Survival Count')
plt.grid(axis='x')
plt.show()
```

## Pie Charts

Pie charts are a popular way to represent parts of a whole, showcasing the proportion of each category as a slice of a circular pie. Each sliceâ€™s size is proportional to its contribution to the total, making pie charts effective for visualizing percentage distributions.

### Features of Pie Charts

1. **Simplicity**:
   - Pie charts provide an easy-to-understand visual representation of data proportions.
   - They are ideal for datasets with a limited number of categories.

2. **Visual Appeal**:
   - The circular layout of pie charts is visually appealing and can effectively highlight differences in data proportions.

### Customization Options

Pie charts offer several customization options to enhance readability and visual impact:

- **Slice Colors**: Different colors can be assigned to each slice to distinguish between categories.
- **Exploding Slices**: Individual slices can be separated from the rest of the pie to highlight specific data points.
- **Labels and Annotations**: Labels can be added to each slice to indicate category names and their corresponding percentages.
- **Legends**: A legend can be included to explain the colors and categories represented in the chart.
- **Percentage Labels**: Percentage values can be displayed on each slice to provide precise information about the distribution.
- **Start Angle**: The starting angle of the first slice can be adjusted to position slices strategically for emphasis.


<a id=41></a>
# <span style='color:#ffffff; font-size:70%;'> Create a basic pie chart.</span>

```python
plt.figure(figsize=(8, 8))
titanic_df['Pclass'].value_counts().plot(kind='pie')
plt.title('Pie Chart: Class Proportion')
plt.ylabel('')
plt.show()
```

<a id=42></a>
# <span style='color:#ffffff; font-size:70%;'> Customize slice colors and add a legend.</span>

```python
plt.figure(figsize=(8, 8))
titanic_df['Pclass'].value_counts().plot(kind='pie', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'], legend=True)
plt.title('Pie Chart with Legend: Class Proportion')
plt.ylabel('')
plt.show()
```

<a id=43></a>
# <span style='color:#ffffff; font-size:80%;'> Explode a slice of the pie chart.</span>

```python
plt.figure(figsize=(8, 8))
titanic_df['Pclass'].value_counts().plot(kind='pie', startangle=90, explode=[0, 0.1, 0], colors=['lightblue', 'lightgreen', 'lightcoral'], legend=True)
plt.title('Exploded Pie Chart: Class Proportion')
plt.ylabel('')
plt.show()
```

<a id=44></a>
# <span style='color:#ffffff; font-size:70%;'> Add percentage labels to slices.</span>

```python
plt.figure(figsize=(8, 8))
titanic_df['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, pctdistance=0.85, colors=['lightblue', 'lightgreen', 'lightcoral'], legend=True)
plt.title('Pie Chart with Percentage Labels: Class Proportion')
plt.ylabel('')
plt.show()
```
